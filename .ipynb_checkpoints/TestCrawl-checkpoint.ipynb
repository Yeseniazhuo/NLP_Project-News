```python
def Download(url):
    '''
    :param url: URL link
    :return: response text
    '''
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/59.0.3071.115 Safari/537.36'}
    response = re.get(url, headers=headers)
    return response.text
```
